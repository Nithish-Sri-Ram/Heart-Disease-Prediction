## Data Processing
After exploreing the dataset, I observer that I need to convert some categorical variables into dummy variables and scale all the values before training the machine learning models. FIrst, I'll use the get_dummies method to create dummy columns for categorical variables.

```python
dataset = pd.get_dummies(df,columns = ['sex','cp','fbs','restecg','exang','slope','ca','thal'])
```

We are applying standard scaling because the features have significant variations and are measured in different units. Standard scaling ensures that all features are rescaled to have a mean of 0 and a standard deviation of 1, bringing them to the same distribution. This helps in balancing the influence of each feature, especially for distance-based algorithms like KNN or SVM, where differences in scales can affect the performance.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
standardScaler = StandardScaler()
columns_to_scale = ['age','trestbps','chol','thalach','oldpeak']
dataset[columns_to_scale]=standardScaler.fit_transform(dataset[columns_to_scale])
```

```python
dataset.head()
```

Now I'm creating my dependent(x) and independent(y) features 

```python
y=dataset['target']                        #Dependent feature
X=dataset.drop(['target'],axis=1)          #Independent features
```

I'm evaluating K-Nearest Neighbors (KNN) for 20 different values of k (from 1 to 20). For each k, I'm using 10-fold cross-validation, where the dataset is split into 10 parts, and the model is trained and validated 10 times. After running these 10 experiments for each k, I calculate the average accuracy score. This process helps determine which value of k provides the best model performance by averaging the cross-validation scores.

#### Steps:
Initialize an empty list knn_scores to store the mean cross-validation scores for each value of k.

Loop through values of k from 1 to 20 (i.e., for k in range(1, 21)):

For each k, create a K-Nearest Neighbors classifier (KNeighborsClassifier) with n_neighbors=k.
Perform 10-fold cross-validation (cross_val_score) on the KNN classifier using the feature matrix X and target vector y.

Set cv=10 to ensure 10-fold cross-validation.
Store the resulting scores in a variable score.
Calculate the mean cross-validation score for the current value of k using score.mean(), and append it to knn_scores.

Repeat steps 2-4 for each value of k from 1 to 20.

Return the list knn_scores containing the mean cross-validation scores for each k.

#### Output:
A list of mean cross-validation scores for each value of k (neighbors)

```python
from sklearn.model_selection import cross_val_score  #This is for cross validations
knn_scores = []
for k in range(1,21):
    knn_classifier = KNeighborsClassifier(n_neighbors = k)
    score = cross_val_score(knn_classifier, X, y,cv=10)
    knn_scores.append(score.mean())
```

We are ploting the accuracy obtained from applying the knn algo at each different locations from 1 to 20

```python
plt.figure(figsize=(14,8))
plt.plot([k for k in range(1,21)],knn_scores, color = 'red')
for i in range(1,21):
    plt.text(i,knn_scores[i-1],f"({i},{knn_scores[i-1]:.4f})")
plt.xticks([i for i in range(1,21)])
plt.xlabel('Number of NEighbors (K)')
plt.ylabel('Scores')
plt.title('K Neighbors Classifier scores for different K values')
```

From the above we see that the k value when it is 12 is having the highest accuracy so we are directly using it to find the accuracy

```python
# Training the final model with k=12 (best accuracy)
knn_classifier = KNeighborsClassifier(n_neighbors = 12)
score=cross_val_score(knn_classifier,X,y,cv=10)
```
